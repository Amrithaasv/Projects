```python
#1 Histogram of Gender Inequality Index (GII)

import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.hist(df['GII'], bins=20, edgecolor='k', color='skyblue', alpha=0.7)
plt.xlabel('Gender Inequality Index (GII)')
plt.ylabel('Frequency')
plt.title('Distribution of GII')
plt.show()
```

```python
#2 Bar Chart of Top 5 Countries with Highest GII

top_5_gii_countries = df.nlargest(5, 'GII')
plt.figure(figsize=(10, 6))
plt.bar(top_5_gii_countries['Country'], top_5_gii_countries['GII'], color='salmon')
plt.xlabel('Country')
plt.ylabel('Gender Inequality Index (GII)')
plt.title('Top 5 Least Performers GII')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#3 ake care of headings
```

```python
bottom_5_gii_countries = df.nsmallest(5, 'GII')
plt.figure(figsize=(10, 6))
plt.bar(bottom_5_gii_countries['Country'], bottom_5_gii_countries['GII'], color='magenta')
plt.xlabel('Country')
plt.ylabel('Gender Inequality Index (GII)')
plt.title('Top 5 Perfomer Countries GII')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#4Scatter Plot of GII vs. Human Development Index (HDI)

plt.figure(figsize=(8, 6))
plt.scatter(df['Human_development'], df['GII'], alpha=0.5, color='purple')
plt.xlabel('Human Development Index (HDI)')
plt.ylabel('Gender Inequality Index (GII)')
plt.title('GII vs. HDI')
plt.show()
```

```python
#5Box Plot of Maternal Mortality Rate

import seaborn as sns

plt.figure(figsize=(8, 6))
sns.boxplot(x=df['Maternal_mortality'], color='teal')
plt.xlabel('Maternal Mortality Rate')
plt.title('Distribution of Maternal Mortality Rate')
plt.show()
```

```python
#6 What is the relationship between Adolescent Birth Rate and GII?


plt.scatter(df['Maternal_mortality'], df['GII'], alpha=0.5, color='purple')
plt.xlabel('Maternal Mortality Rate')
plt.ylabel('Gender Inequality Index (GII)')
plt.title('Maternal Mortality Rate vs. GII')
plt.show()
```

```python
#7
plt.scatter(df['Maternal_mortality'], df['F_Labour_force'], alpha=0.5)
plt.xlabel('Maternal Mortality Rate')
plt.ylabel('Female Labour Force Participation')
plt.title('Maternal Mortality vs. Female Labour Force Participation')
plt.show()
```

```python
#8 What is the relationship between Adolescent Birth Rate and GII?


plt.scatter(df['Adolescent_birth_rate'], df['GII'], alpha=0.5, color='purple')
plt.xlabel('Adolescent Birth Rate')
plt.ylabel('Gender Inequality Index (GII)')
plt.title('Adolescent Birth Rate vs. GII')
plt.show()
```

```python
#9Selecting relevant columns (excluding GII and HDI)
columns_of_interest = ['Maternal_mortality', 'Adolescent_birth_rate', 'Seats_parliament',
                        'F_secondary_educ', 'M_secondary_educ', 'F_Labour_force', 'M_Labour_force']

# Calculate the correlation coefficients
correlation_coefficients = df[columns_of_interest].corrwith(df['GII'])

# Find the column with the highest absolute correlation coefficient
most_contributing_column = correlation_coefficients.abs().idxmax()

# Print the result
print(f"The column that contributes most to GII, other than GII and HDI, is: {most_contributing_column}")
```

```python
#Scatter Plot Matrix of GII and Other Variables

sns.pairplot(df[['GII', 'Maternal_mortality', 'F_Labour_force', 'Human_development']])
plt.show()
```

```python
#Violin Plot of GII by HDI Category

plt.figure(figsize=(10, 6))
sns.violinplot(x='Human_development', y='GII', data=df, palette='viridis')
plt.xlabel('HDI Category')
plt.ylabel('Gender Inequality Index (GII)')
plt.title('GII by HDI Category')
plt.tight_layout()
plt.show()
```
