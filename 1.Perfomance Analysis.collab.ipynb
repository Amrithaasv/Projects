````python
import seaborn as sns
import matplotlib.pyplot as plt
````

````python
df = pd.ExcelFile('Data set for BA (2).xlsx')
sheet_names = df.sheet_names
sheet_names
````

````python
# Create a bar chart for new users and engaged sessions by channel group

plt.figure(figsize=(12, 8))
sns.barplot(x='First user default channel group', y='New users', data=df_user_acquisition, color='skyblue', label='New Users')
sns.barplot(x='First user default channel group', y='Engaged sessions', data=df_user_acquisition, color='orange', label='Engaged Sessions')
plt.xlabel('Channel Group')
plt.ylabel('Count')
plt.title('New Users vs Engaged Sessions by Channel Group')
plt.xticks(rotation=45)
plt.legend()
plt.show()
````

New users from Organic Search category has maximum number of engagement sessions.


````python
# Create a bar chart for new users and engagement rate by channel group
plt.figure(figsize=(12, 8))
sns.barplot(x='First user default channel group', y='New users', data=df_user_acquisition, color='skyblue', label='New Users')
sns.barplot(x='First user default channel group', y='Engagement rate', data=df_user_acquisition, color='orange', label='Engagement Rate')
plt.xlabel('Channel Group')
plt.ylabel('Count')
plt.title('New Users vs Engagement Rate by Channel Group')
plt.xticks(rotation=45)
plt.legend()
plt.show()
````

New users from Display search are having maximum engagement rate.
Least engagement rate comes from users who came from Organic social search.
Paid Search users are not contributing muchwhich is an issue

The correlation coefficients between 'Users', 'New users', 'Engaged sessions', and 'Event count' are extremely close to 1.
This suggests that as one of these variables increases, the others tend to increase as well.

The correlation coefficients between the other variables and 'Engagement rate' are very close to 0.The weak correlation suggests that changes in the engagement rate may not strongly coincide with changes in the number of users, new users, or engaged sessions.
